```python
count = ['airline','source_city','departure_time','stops','arrival_time','destination_city','class']
colors = ['blue', 'green', 'orange', 'purple', 'red', 'brown', 'k']
for i, color in zip(count,colors):
  sorted_counts = df[i].value_counts(ascending=True).index
  a = sns.countplot(x=i,data=df,order=sorted_counts,color=color)
  plt.title('Count Graph of '+ i,color=color)
  a.set_xlabel(i, fontsize=12,color=color)  # Set x-label size
  a.set_ylabel('Count', fontsize=12,color=color)
  for j in a.containers:
    a.bar_label(j)
  plt.show()
  print("--------------------------------------------------------------------------------------------------------------")
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
df['class'] = encode.fit_transform(df['class'])
```

```python
df1 = pd.get_dummies(df[['airline','source_city','departure_time','stops','arrival_time','destination_city']],dtype=int,drop_first=True)
df1
```

```python
new = pd.concat([df,df1],axis=1)
new
```

```python
new.drop(['Unnamed: 0','airline','flight','source_city','departure_time','stops','arrival_time','destination_city'],axis=1,inplace=True)
new
```

```python
x = new.drop(['price'],axis=1)
y = new['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=0)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
mape = mean_absolute_percentage_error(y_test,y_pred)
mape
```

```python
score = r2_score(y_test,y_pred)
score
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
x_poly
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(x_poly,y)
y_poly = model1.predict(x_poly)
y_poly
```

```python
mape1 = mean_absolute_percentage_error(y,y_poly)
mape1
```

```python
score1 = r2_score(y,y_poly)
score1
```
